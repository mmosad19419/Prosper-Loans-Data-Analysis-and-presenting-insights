```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("./prosper_loans_master.csv")
```

```python
base_color = sns.color_palette()[0]
def boxplot(df, variable1, variable2, ticks_rotation, title, xlabel, ylabel):
    plt.figure(figsize=(15,10))
    sns.boxplot(data = df, x = variable1, y = variable2, color = base_color)
    plt.xticks(rotation = ticks_rotation);
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    
# define customized pointplot
def pointplot(df, variable1, variable2, hue, ticks_rotation, title, xlabel, ylabel):
    plt.figure(figsize=(15,10))
    sns.pointplot(data = df, x = variable1, y = variable2, hue = hue, errorbar = ("ci", 95))
    plt.xticks(rotation = ticks_rotation);
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

__What factors affect a defaulted loans loan’s outcome status?__

-  Deafaulted Loans has outliers in Prosper score, we will see a summary statistics of Deafaulted Loans Prosper score

-  past due loans has the lowest prosper score with charged off

-  defaulted loans amount average is the same as completed loans amount, means we can not depend on the loan amount to predict its final status.

- Deafaulted Loans has outliers in EstimatedEffectiveYield, we will see a summary statistics of Deafaulted Loans EstimatedEffectiveYield

- part time borrower has high percentage of outliers in the prosper score

- Employed people has the highest prosper score

- part time employment get the lowest loan amounts which relate to the low prosper score, on the other hand employment get the highest

- In the estimatedeffectiveyield the part time employment has high percentage of outliers, Employed borrowers has the highest estimatedeffectiveyield

```python
# explore the relation between "loan status" and "Prosper score"
boxplot(df, "loanstatus","prosperscore", 15, "Summary of prosper score for each loan status", "Loan Status", "Prosper Score")
```

```python
# explore the relation between "loan status" and "EstimatedEffectiveYield"
boxplot(df, "loanstatus","estimatedeffectiveyield", 15, "Summary of EstimatedEffectiveYield for each loan status", "Loan Status", "EstimatedEffectiveYield")
```

```python
deafulted_loans_prosperscore = pd.DataFrame(df[df.loanstatus == "Defaulted"].prosperscore.describe()).reset_index()
deafulted_loans_prosperscore
deafulted_loans_prosperscore.drop([0], axis = 0, inplace = True)
# plot the descriptive summary statistics
plt.figure(figsize=(9, 6))
sns.scatterplot(data = deafulted_loans_prosperscore, x = "index", y = "prosperscore", marker = "x",s = 400)
plt.xlabel("Prosper Score Summary Values")
plt.ylabel("Statistics Summary Variables")
plt.title("Summary Statistics For Defaulted loans Prosper Score Values ")
plt.grid()
plt.show()
```

```python
deafulted_loans_estimatedeffectiveyield = pd.DataFrame(df[df.loanstatus == "Defaulted"].estimatedeffectiveyield.describe())
deafulted_loans_estimatedeffectiveyield.reset_index(inplace = True)
deafulted_loans_estimatedeffectiveyield.drop([0], axis = 0, inplace = True)

# plot the descriptive summary statistics
plt.figure(figsize=(9, 6))
sns.scatterplot(data = deafulted_loans_estimatedeffectiveyield, x = "index", y = "estimatedeffectiveyield", marker = "x",s = 400)
plt.xlabel("EstimatedEffectiveYield Summary Values")
plt.ylabel("Statistics Summary Variables")
plt.title("Summary Statistics For Defaulted loans EstimatedEffectiveYield Values ")
plt.grid()
plt.show()
```

### What is the preferred loan term in each listing category?
